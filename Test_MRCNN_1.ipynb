```python
# 1. Install micromamba dengan path yang benar
!wget -qO- https://micro.mamba.pm/api/micromamba/linux-64/latest | tar -xvj bin/micromamba

# 2. Buat environment python 3.7
!./bin/micromamba create -y -p /usr/local/env_maskrcnn python=3.7 pip

# 3. Upgrade pip dan install dependencies dasar
!./bin/micromamba run -p /usr/local/env_maskrcnn pip install --upgrade pip setuptools wheel

# 4. Install dependencies dengan versi yang kompatibel
!./bin/micromamba run -p /usr/local/env_maskrcnn pip install \
    numpy==1.19.5 \
    opencv-python-headless==4.5.5.64 \
    pillow==8.3.2 \
    scikit-image==0.17.2 \
    matplotlib==3.3.4 \
    cython==0.29.24 \
    h5py==2.10.0 \
    imgaug==0.4.0 \
    pycocotools==2.0.6

# 5. Install TensorFlow + Keras yang benar-benar kompatibel
!./bin/micromamba run -p /usr/local/env_maskrcnn pip install \
    tensorflow==1.15.5 \
    keras==2.3.1 \
    protobuf==3.19.6

# 6. Install dependencies tambahan yang diperlukan
!./bin/micromamba run -p /usr/local/env_maskrcnn pip install \
    scipy==1.5.4 \
    pandas==1.1.5 \
    IPython

print("âœ… Environment setup completed!")
```
